```python
# ==========================
# TITANIC MINI DASHBOARD
# Task 4 - Maincrafts
# ==========================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8,5)

# Load Dataset from URL
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(titanic_url)

print("="*50)
print("First Five Rows")
print(df.head())

print("="*50)
print("Dataset Shape")
print(df.shape)

print("="*50)
print("Dataset Information")
print(df.info())

print("="*50)
print("Missing Values")
print(df.isnull().sum())

# -----------------------------
# Data Cleaning
# -----------------------------

# Fill Age with Mean
df["Age"].fillna(df["Age"].mean(), inplace=True)

# Fill Embarked with Mode
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

# Drop Cabin
df.drop(columns=["Cabin"], inplace=True)

print("\nMissing Values After Cleaning")
print(df.isnull().sum())

# -----------------------------
# Feature Engineering
# -----------------------------

# Family Size
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

# Age Groups
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0,12,18,35,60,100],
    labels=["Child","Teen","Adult","Middle Age","Senior"]
)

print("\nDataset Ready")
print(df.head())
# ==========================================
# 1. Survival Count
# ==========================================

plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Passengers")
plt.show()

print("Insight:")
print("More passengers died than survived in the Titanic disaster.\n")


# ==========================================
# 2. Survival by Gender
# ==========================================

plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()

print("Insight:")
print("Female passengers had a much higher survival rate than male passengers.\n")


# ==========================================
# 3. Survival by Passenger Class
# ==========================================

plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()

print("Insight:")
print("Passengers travelling in First Class survived more frequently than those in lower classes.\n")


# ==========================================
# 4. Age Distribution
# ==========================================

plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()

print("Insight:")
print("Most passengers were young adults between 20 and 40 years of age.\n")


# ==========================================
# 5. Fare Distribution
# ==========================================

plt.figure(figsize=(8,5))
sns.histplot(df["Fare"], bins=40, kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.show()

print("Insight:")
print("Most passengers paid lower fares, while only a few paid very high ticket prices.\n")
# ==========================================
# 6. Correlation Heatmap
# ==========================================

plt.figure(figsize=(10,6))

corr = df.select_dtypes(include=np.number).corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()

print("Insight:")
print("Passenger Class and Fare show a strong relationship, while Survival has moderate correlation with Fare and Pclass.\n")


# ==========================================
# 7. Survival by Embarkation Port
# ==========================================

plt.figure(figsize=(6,4))

sns.countplot(
    x="Embarked",
    hue="Survived",
    data=df
)

plt.title("Survival by Embarkation Port")
plt.show()

print("Insight:")
print("Passengers boarding from Cherbourg (C) had the highest survival rate.\n")


# ==========================================
# 8. Survival by Family Size
# ==========================================

plt.figure(figsize=(8,5))

sns.barplot(
    x="FamilySize",
    y="Survived",
    data=df,
    estimator=np.mean
)

plt.title("Survival Rate by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Average Survival Rate")
plt.show()

print("Insight:")
print("Passengers travelling with small families generally had better survival chances than those travelling alone or in very large families.\n")


# ==========================================
# 9. Age vs Survival (Box Plot)
# ==========================================

plt.figure(figsize=(7,5))

sns.boxplot(
    x="Survived",
    y="Age",
    data=df
)

plt.title("Age Distribution by Survival")
plt.xlabel("Survived")
plt.ylabel("Age")
plt.show()

print("Insight:")
print("Children and younger passengers generally had better survival chances than older passengers.\n")


# ==========================================
# 10. Passenger Class vs Gender
# ==========================================

plt.figure(figsize=(7,5))

sns.countplot(
    x="Pclass",
    hue="Sex",
    data=df
)

plt.title("Passenger Class by Gender")
plt.show()

print("Insight:")
print("Most male passengers travelled in Third Class, while First Class had a relatively balanced gender distribution.\n")
```
